```python
Session_Date=[]

start_time=[]
stop_time=[]
schedule_Start_Time=[]
schedule_End_Time=[]


program_name=[]
session_type=[]
session_term=[]

year_no=[]
level_no=[]
course_name=[]
course_code=[]
session_mode=[]
subject_name=[]
Primary_staff=[]
sub_staff=[]


Total_student=[]
auto_a=[]
manual_a=[]
auto_p=[]
manual_p=[]


duration_as_per_zoom=[]
infra_name=[]
title=[]
sub_type=[]
merge_ID=[]
auto_persentage=[]
primary_present=[]
primary_persentage=[]
```

```python
for no ,session in df.iterrows():
    start=int(session["sessionDetail"]["start_time"]["$date"]["$numberLong"])
    start = datetime.fromtimestamp(int(start)/1000)
    
    #date
    Session_Date.append(start.strftime("%Y-%m-%d"))
    
    #session started time
    start_time.append(start.strftime("%H:%M"))

    try:
        stop=session["sessionDetail"]["stop_time"]["$date"]["$numberLong"]
        stop = datetime.fromtimestamp(int(stop)/1000)
        
        #session ended time
        stop_time.append(stop.strftime("%H:%M"))
    except:
        stop=session["sessionDetail"]["stop_time"]["$numberLong"]
        stop = datetime.fromtimestamp(int(stop)/1000)
        
        #session ended time
        stop_time.append(stop.strftime("%H:%M"))
        
 
    schedule_Start=session["scheduleStartDateAndTime"]["$date"]["$numberLong"]
    schedule_Start = datetime.fromtimestamp(int(schedule_Start)/1000)
    
    #session starting time as per schedule time
    schedule_Start_Time.append(schedule_Start.strftime("%H:%M"))
    
    
    schedule_end=session["scheduleEndDateAndTime"]["$date"]["$numberLong"]
    schedule_end = datetime.fromtimestamp(int(schedule_end)/1000)
    
    #session ending time as per schedule time
    schedule_End_Time.append(schedule_end.strftime("%H:%M"))
    
    #program_name
    program_name.append(session["program_name"])
    
    #session_type
    session_type.append(session["type"])
    
    #session_term
    session_term.append(session["term"])
    
    #year_no
    year_no.append(session["year_no"])
    
    #level_no
    level_no.append(session["level_no"])
    
    
    #course_name
    course_name.append(session["course_name"])
    
    
    #course_code
    course_code.append(session["course_code"])
    
    #session_mode
    session_mode.append(session["mode"])
    
    #subject_name
    subject_name.append(session["subjects"][0]["subject_name"])
    
    
    pri_staff_id=(session["sessionDetail"]['startBy']['$oid'])
    if len(session['staffs'])==1:
        f_n=session['staffs'][0]['staff_name']['first']
        l_n=session['staffs'][0]['staff_name']['last']
        name=(f_n+" "+l_n)
        
        #Primary_staff
        Primary_staff.append(name)
        
        #sub_staff
        sub_staff.append("-")
    else:
        sub_staffs=[]
        for staf in session['staffs']:
            if staf['_staff_id']['$oid'] == pri_staff_id:
                f_n=staf['staff_name']['first']
                l_n=staf['staff_name']['last']
                name=(f_n+" "+l_n)
                
                #Primary_staff
                Primary_staff.append(name)
            else:
                f_n=staf['staff_name']['first']
                l_n=staf['staff_name']['last']
                name=(f_n+" "+l_n)  
                sub_staffs.append(name)
                
        #sub_staff        
        sub_staff.append(",".join(sub_staffs))
        
    t_student=0 
    auto_a_=0
    manual_a_=0
    auto_p_=0
    manual_p_=0
    
    
    pri_present_=0
    no_pri_=0
    pending_=0
    for stu in session['students']:
        
        
        
        
        try:
            if stu['primaryStatus']=="present":
                pri_present_+=1
            elif stu['primaryStatus']=="pending":
                pending_+=1
        except:
            no_pri_+=1
        
        t_student+=1
        if stu['status'] == "present":
            try:
                if stu['mode']=='auto':
                    auto_p_+=1
                else:
                    manual_p_+=1
            except:
                auto_p_+=1
            
        elif stu['status'] == "absent":
            try:
                if stu['mode']=='auto':
                    auto_a_+=1
                else:
                    manual_a_+=1
            except:
                auto_a_+=1
    #Total_student
    
    total_pre=auto_p_+manual_p_
    if no_pri_ >0:
        primary_present.append('-')
        primary_persentage.append('-')
    elif pending_ >0:
        primary_present.append('pending')
        primary_persentage.append('pending')
    else:
        primary_present.append(pri_present_)
        try:
            Pri_per=round((pri_present_/total_pre)*100)
            primary_persentage.append(str(Pri_per)+" %")
        except:
            primary_persentage.append("0 %")
    Total_student.append(t_student)
    auto_a.append(auto_a_)
    manual_a.append(manual_a_)
    auto_p.append(auto_p_)
    manual_p.append(manual_p_)
    
    if t_student != 0:
        per=round(((auto_p_+auto_a_)/t_student)*100)
        per=str(per)+" %"
    else:
        per="0 %"
    auto_persentage.append(per)
    #infra_name
    infra_name.append(session['infra_name'])
    
    #title
    title.append(session['title'])
    
    #sub_type
    sub_type.append(session['sub_type'])
    
   # print(session['staffs'])
    try:
        time_dur=session['staffs'][0]['duration']
        time_dur=round(int(time_dur)/60)
        duration_as_per_zoom.append(time_dur)
    except:
        duration_as_per_zoom.append("-")
    if session['merge_status'] == True:
        merge_ID.append(session["merge_with"][0]['session_id']['$oid'])
    else:
        merge_ID.append("")
```

```python
df.replace(np.nan, "NaN", inplace=True)

# Find the duplicate rows based on row-wise comparison
duplicates = df.duplicated(keep=False)

# Count the occurrences of each duplicate row
duplicate_counts = df[duplicates].groupby(df.columns.tolist()).cumcount() + 1

# Add a new column 'Duplicates' and assign the count + 1 to all duplicate rows
df['Merged_count'] =duplicate_counts

# Merge the duplicate rows and keep the first occurrence of each group
subset_columns = [col for col in df.columns if col != 'Merged_count']

df_merged = df.drop_duplicates(subset=subset_columns, keep='last')
df_merged['Merged_count'] = df_merged['Merged_count'].replace(np.nan,0)
df_merged['Merged_count'] = df_merged['Merged_count'].astype(int)
```
